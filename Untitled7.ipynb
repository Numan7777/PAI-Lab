```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

# Step 1: Define Variables and Dataset [cite: 5]
# Data points extracted from the experiment source [cite: 41, 42]
data = pd.DataFrame(data={
    'Rain': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'TrafficJam': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'],
    'ArriveLate': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
})

# Step 2: Define Bayesian Network Structure [cite: 8]
# Structure: Rain -> TrafficJam -> ArriveLate [cite: 12, 13, 44]
model = BayesianNetwork([('Rain', 'TrafficJam'), ('TrafficJam', 'ArriveLate')])

# Step 3: Parameter Learning [cite: 16]
# Apply Maximum Likelihood Estimation (MLE) to learn the CPDs 
model.fit(data, estimator=MaximumLikelihoodEstimator)

# Step 4: Display Conditional Probability Distributions [cite: 21]
print("--- Learned CPDs ---")
for cpd in model.get_cpds():
    print(cpd) [cite: 47, 53]

# Step 5: Initialize Inference Engine [cite: 23]
# Use Variable Elimination for exact probabilistic inference [cite: 24, 49]
inference = VariableElimination(model)

# Step 6: Perform Probabilistic Query [cite: 26]
# Compute posterior probability: P(ArriveLate | Rain = Yes) [cite: 28, 29, 51]
print("\n--- Inference Result ---")
query_result = inference.query(variables=['ArriveLate'], evidence={'Rain': 'Yes'})

# Step 7: Display Results [cite: 31]
print(query_result) [cite: 51, 60]
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

data = pd.DataFrame(data={
    'Rain': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'TrafficJam': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'],
    'ArriveLate': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
})

model = DiscreteBayesianNetwork([('Rain', 'TrafficJam'), ('TrafficJam', 'ArriveLate')])

model.fit(data, estimator=MaximumLikelihoodEstimator)

for cpd in model.get_cpds():
    print(cpd)

inference = VariableElimination(model)

query_result = inference.query(variables=['ArriveLate'], evidence={'Rain': 'Yes'})
print(query_result)
```
