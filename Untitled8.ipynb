```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
import pandas as pd

data = pd.DataFrame(data={
    'IncomeStability': ['High', 'Low', 'Low', 'High', 'Low', 'High', 'High', 'Low'],
    'EmploymentType': ['Salaried', 'Self', 'Self', 'Salaried', 'Self', 'Salaried', 'Self', 'Salaried'],
    'CreditHistory': ['Good', 'Bad', 'Bad', 'Good', 'Bad', 'Good', 'Bad', 'Good'],
    'DefaultRisk': ['No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
})

model = DiscreteBayesianNetwork([
    ('IncomeStability', 'DefaultRisk'),
    ('EmploymentType', 'DefaultRisk'),
    ('CreditHistory', 'DefaultRisk')
])

model.fit(data, estimator=MaximumLikelihoodEstimator)

print("--- Learned Conditional Probability Distributions ---")
for cpd in model.get_cpds():
    print(cpd)

inference = VariableElimination(model)

print("\n--- Inference: Probability of Default Risk given Low Income and Bad Credit ---")
query_result = inference.query(
    variables=['DefaultRisk'], 
    evidence={'IncomeStability': 'Low', 'CreditHistory': 'Bad'}
)
print(query_result)
```
